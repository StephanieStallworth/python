# Calculate a new column out of existing column

```python
# Create new column without values that satisfy this condition (mask() means without) 
# Replacing all those values with NaN values
# So Pandas knows not to do calculations on those values 
import numpy as np 
df["TG0"] = df['   TG'].mask(df['   TG']==-9999, np.nan)
df
```

```python
df["TG"] = df['   TG']/10
```

```python
df
```

```python
df["Fahrenheit"] = df["TG"] * (9/5) + 32
```
