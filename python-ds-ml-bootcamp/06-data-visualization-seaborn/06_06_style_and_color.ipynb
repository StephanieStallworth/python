# Style and Color

We've shown a few times how to control figure aesthetics in seaborn, but let's now go over it formally:

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

## Styles

You can set particular styles:

```python
# Basic countplot 
sns.countplot(x='sex',data=tips)
```

```python
# Call set_style() 
# Can pass in arguments: white, ticks, darkgrid, or whitegrid 
sns.set_style('white')
sns.countplot(x='sex',data=tips)
```

```python
# Set style to ticks to better visualize 
sns.set_style('ticks')
sns.countplot(x='sex',data=tips,palette='deep')
```

## Spine Removal

```python
# despine() with no arguments will remove top and right spines by default
sns.countplot(x='sex',data=tips)
sns.despine()
```

```python
# To continue removing bottom and left spine 
# Pass them as arguments to despine()
# Options are top, right, left, and bottom
sns.countplot(x='sex',data=tips)
sns.despine(left=True, bottom=True)
```

## Size and Aspect

You can use matplotlib's **plt.figure(figsize=(width,height) ** to change the size of most seaborn plots.

You can control the size and aspect ratio of most seaborn grid plots by passing in parameters: size, and aspect. For example:

```python
# Non Grid Plot
# Note: set_style stays in place until you set a new style
plt.figure(figsize=(12,3))

# So when seaborn calls matplotlib
# Matplotlib will set the figure size for the seaborn plot
sns.countplot(x='sex',data=tips)
```

```python
# Grid Type Plot
sns.lmplot(x='total_bill',y='tip',size=2,aspect=4,data=tips)
```

## Scale and Context

The set_context() allows you to override default parameters:

```python
# Call set_context()
# Takes in context argument: paper, notebook (default), talk, poster  
# Pass in font_scale argument to increase font size  
sns.set_context('poster',font_scale=4)
sns.countplot(x='sex',data=tips,palette='coolwarm')
```
