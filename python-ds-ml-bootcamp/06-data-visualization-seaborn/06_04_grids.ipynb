# Grids

Grids are general types of plots that allow you to map plot types to rows and columns of a grid, this helps you create similar plots separated by features.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

## PairGrid

Pairgrid is a subplot grid for plotting pairwise relationships in a dataset.

```python
# Just the Grid
# Gives us an empty pairgrid 
# Makes the subplots for us,takes all the numerical columns and "grids" them up 
sns.PairGrid(iris)
```

```python
# Then you map to the grid
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
# Map to upper,lower, and diagonal
# Choose what plot types we want to map 
# Passing the plot function themselves
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

## pairplot

pairplot is a simpler version of PairGrid (you'll use quite often)

```python
# Simplified version of PairGrid above
# Does alot of the stuff for us automatically 
# Use this most often because its simplier and gives us most the info we want
# Use PairGrid if you want to do more customization
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species',palette='rainbow')
```

## Facet Grid

FacetGrid is the general way to create grids of plots based off of a feature:

```python
# Load dataset
tips = sns.load_dataset('tips')
```

```python
# View dataset 
tips.head()
```

```python
# Just the Grid
# Pass in the data, then what you want the column and row to be 
g = sns.FacetGrid(tips, col="time", row="smoker")
```

```python
# Create empty FacetGrid
# Then map plot types to this grid 
# Maps plot type and separates results by column and row
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
```

```python
# Plot types that need more than one argument

# Create FacetGrid same as above 
g = sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')

# Add in as a additional argument in the map call 
# Notice how the arguments come after plt.scatter call
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```

## JointGrid

JointGrid is the general version for jointplot() type grids, for a quick example:

```python
# Create empty JointGrid
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
```

```python
# Then pass in the plot type 
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot(sns.regplot, sns.distplot)
```

Reference the documentation as necessary for grid types, but most of the time you'll just use the easier plots discussed earlier.
