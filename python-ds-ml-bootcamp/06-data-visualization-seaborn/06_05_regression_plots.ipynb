# Regression Plots

Seaborn has many built-in capabilities for regression plots, however we won't really discuss regression until the machine learning section of the course, so we will only cover the **lmplot()** function for now.

**lmplot** allows you to display linear models, but it also conveniently allows you to split up those plots based off of features, as well as coloring the hue based off of features.

Let's explore how this works:

```python
# Imports 
import seaborn as sns
%matplotlib inline
```

```python
# Load dataset 
tips = sns.load_dataset('tips')
```

```python
# View dataset  
tips.head()
```

## lmplot()

```python
# Returns scatterplot with linear fit on top of it 
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
# Pass column to hue argument if you want separation based on a categorical feature
# Returns two scatterplots and two linear fits on top of each other
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
# Specify palette to better see Male vs Female   
# Can infer from this plot that male and female have the same linear fit as far as total_bill vs tip 
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

### Working with Markers

lmplot kwargs get passed through to **regplot** which is a more general form of lmplot(). regplot has a scatter_kws parameter that gets passed to plt.scatter. So you want to set the s parameter in that dictionary, which corresponds (a bit confusingly) to the squared markersize. In other words you end up passing a dictionary with the base matplotlib arguments, in this case, s for size of a scatter plot. In general, you probably won't remember this off the top of your head, but instead reference the documentation.

```python
# With lmplot(), Seaborn is calling matplotlib under the hood
# So can pass in matplotlib style parameters 
# Pass in a dictionary with base matplotlib arguments
# http://matplotlib.org/api/markers_api.html
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',
           markers=['o','v'], # symbols come from matplotlib marker symbols
           scatter_kws={'s':100} # affect matplotlib directly from seaborn 
          )
```

## Using a Grid

We can add more variable separation through columns and rows with the use of a grid. Just indicate this with the col or row arguments:

```python
# Instead of separting by hue on the same plot
# Returns two plots separated by the category ("separating by columns")
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
# Can also separate by row  
# Doing the same as grid calls from earlier but more automatic
# Just adding row and column to lmplot() call 
# Easier to do this, don't have to deal with FacetGrid all the time 
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips)
```

```python
# Add hue argument back in to plot more information
# Elminate row to make it easier to read 
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm')
```

## Aspect and Size

Seaborn figures can have their size and aspect ratio adjusted with the **size** and **aspect** parameters:

```python
# Customize aspect size and ratio 
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',
          aspect=0.6, # Ratio of the height and width between each of the separate plots
           size=8 # Size of each plot
          )
```
