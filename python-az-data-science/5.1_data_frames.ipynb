```python
# Awesome! Let's use this in practice now:
Filter2 = stats.BirthRate > 40
```

```python
# Don't have to create the filter as a separate object 
# Instead of: stats[Filter2]

# Pass in filter directly: 
stats[stats.BirthRate > 40]
```

```python
# Even more interesting:
stats[stats.BirthRate > 40 and stats.InternetUsers < 2] # Error. Why?
```

```python
# Error is here
# Python is not a vectorized programming language
# By default, what this "and" means its trying to work with two single values
# Its waiting for a single True or False value on the left side
# And a single True or False value on the right side
# Then will combine them into either a True or a False
# So it doesn't know how to work with objects that have lots of values in them which is what we have here
# These are series on either side, not single values
# "ValueError: The truth value of a Series is ambiguous"

# stats.BirthRate > 40 and stats.InternetUsers < 2     # Error. Why?!

# For simplicity use object names
Filter and Filter2       # Error. Why?!
```

```python
# Use bit-wise ampersand to compare element by element
# Only when both conditions are met we'll get a True
Filter & Filter
```

```python
stats[Filter & Filter2]
```

```python
# Have to specify order of operations to avoid error
stats[(stats.BirthRate > 40) & (stats.InternetUsers < 2)]
```

```python
############## Visualizating with Seaborn: Part 1 #######################
# lmplot = linear model plot without regression line 
# If you specify names of parameters, doesn't matter which order you specify them in  
vis3 = sns.lmplot(data = stats, x='InternetUsers', y='BirthRate', fit_reg = False)
```

```python
vis4 = sns.distplot(stats["BirthRate"], bins=30)
```

```python
# Add color by group 
# Ignores plt.rcParams['figure.figsize'] = 8, 4 
# Because it fits across a facet grid 
# Pass the size and aspect arguments to modify  
vis3 = sns.lmplot(data = stats, x='InternetUsers', y='BirthRate', \
                  fit_reg = False, hue='IncomeGroup', size = 10)
```

```python
############ Keyword Arguments  in Python (Advanced Tutorial) ############

# Marker Size:
# https://seaborn.pydata.org/generated/seaborn.lmplot.html
# https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.html

# seaborn.lmplot is a wrapper for matplotlib.pyplot
# matplotlib.pyplot has a parameter "s" (for size)
# seaborn.lmplot did not include this as a parameter
# But will still go through if you pass it as the dictionary: scatter_kws
# scatter_kws will forward on those parameters to plt.scatter
vis3 = sns.lmplot(data = stats, x='InternetUsers', y='BirthRate', \
                  fit_reg = False, hue='IncomeGroup', size = 10, \
                 scatter_kws={"s":100})
```
